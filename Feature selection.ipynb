# House price prediction

## Predicting the price of houses in Washnington DC in USA.

### Feature selection

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
import pandas as pd
import numpy as np
```

```python
data_frame=pd.read_csv('new_data.csv')
data_frame.head()
```

```python
x=data_frame.iloc[:,:-1]
y=data_frame[['price']]
```

```python
model=SelectFromModel(Lasso(random_state=0,alpha=0.005))
```

```python
model.fit(x,y)
```

```python
model.get_support()
```

```python
selected_features=[]
for i,j in zip(model.get_support(),x.columns):
    if i==True:
        selected_features.append(j)
```

```python
selected_features
```

```python
final_data=pd.concat([data_frame[selected_features],data_frame[['price']].reset_index(drop=False)],axis=1)
```

```python
final_data.head()
```

```python
final_data.to_csv('final_data.csv',index=False)
```
